### Model evaluation and improvement




Để đánh giá chất lượng  của một mô hình chúng ta phân  chia dữ liệu thành hai phần bằng hàm **train_test_split()** trong đó mô hình được xây dựng  trên  bộ dữ liệu huấn luyện bằng hàm **fit()** và mô hình được đánh giá trên bộ dữ liệu kiểm định với hàm **score()**. Xét một ví dụ đơn giản sau: 



```python
# Load gói pandas và đọc vào dữ liệu: 
import pandas as pd
credit = pd.read_csv("C:/Users/win10ls/Desktop/Data_Science/credit.csv")

# Xem qua dữ liệu: 
credit.shape
```

```python
# Một số quan sát đầu: 
credit.head()
```

```python
# Loại cột biến id: 
credit.drop("id", axis = 1, inplace = True)
credit.head()
```

```python
# Hồ sơ được đánh số 1 = Default và 0 = NonDefault: 
credit.Y.unique()
```

```python
# Dán lại nhãn cho cột biến Y: 
Y = credit["Y"]
Y[Y == 1] = "B"
Y[Y == 0] = "G"
# Thay thế Y mới ở trên: 
credit = credit.assign(Y = Y)
```

```python
credit.head()
```

```python
y = Y
y.head()
```

```python
X = credit.drop("Y", 1)
X.head()
```

```python
# Phân chia dữ liệu theo tỉ lệ 50 - 50: 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state = 29)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
# Thực hiện hồi quy Logistic: 
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression().fit(X_train, y_train)

# Đánh giá chất lượng của  mô hình dự báo: 
logreg.score(X_test, y_test)
```

```python
# So sánh để thấy tỉ lệ dự báo này trên mẫu thử nghiệm là không cao đáng kể so với đoán ngẫu nhiên: 
sum(Y == "G") / len(y)
```

```python
# Thực hiện cross-validation (mặc định k = 3) với đảm  bảo là phân phối của
# biến  đích không đổi (stratifed k-fold cross-validation)
from sklearn.model_selection import cross_val_score
logreg = LogisticRegression()
scores = cross_val_score(logreg, X, Y)
print("Cross-validation accuracy: ", scores)
```

```python
# Chọn k = 5: 
scores5 = cross_val_score(logreg, X, Y, cv = 5)
scores5
```

```python
#  Tính mean: 
scores5.mean()
```

```python
# Một cách khác với lựa chọn shuffle = True để khắc phục vấn dề phân phối của biến đích:  
from sklearn.model_selection import KFold
kfold = KFold(n_splits = 5, random_state = 9, shuffle = True)
cross_val_score(logreg, X, Y, cv = kfold)
```
